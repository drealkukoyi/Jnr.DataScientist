```python
# Total number of job positions to check
total_positions = 3
position_index = 0

# Minimum years of experience required for each data science job position
position_experience_requirements = [1, 3, 2]
```

```python
while position_index < total_positions:
    print('Position requires', position_experience_requirements[position_index], 'years of experience.')
    position_index += 1
```

```python
# User's years of experience
user_years_of_experience = 2
```

```python
while position_index < total_positions:
    required_years = position_experience_requirements[position_index]
    if user_years_of_experience >= required_years:
        print('Qualified')
    else:
        print('Not Qualified')
        position_index += 1
```

```python
for x in position_experience_requirements:
    print(f'Position requires {x} years of experience.')
```

**`Enumerate`**

```python
for index, years in enumerate(position_experience_requirements, start=1):
    print(f'Position {index} requires {years} years of experience.')
```

**`Pass`**

```python
for x in position_experience_requirements:
    pass
```

**Nested Loop**

```python
roles = ['Data Scientist', 'Machine Learning Engineer']
skills = ['Python', 'SQL', 'Machine Learning']

for role in roles:
    print(f'For the role of {role}, you need experience in:')
    for skill in skills:
        print(f' - {skill}')
```

```python
# For loop to check qualification for each job
for required_experience in position_experience_requirements:
    if user_years_of_experience >=  required_experience:
        print('Qualified')
    else:
        print('Not Qualified')
```

**`Break`**

```python
for required_experience in position_experience_requirements:
    if user_years_of_experience >= required_experience:
        break
    else:
        print('Not Qualified')
```

#### OUTLIER DETECTION AND REMOVAL USING 3 STANDARD DEVIATION

```python
import pandas as pd
```

```python
df = pd.read_csv('height.csv')
df.sample(5)
```

This means removing 3 standard deviations from the mean. Therefore, we can come up with upper and lower bounds.


```python
df.Height.mean()
```

```python
df.Height.std()
```

```python
upper_limit = df.Height.mean() + 3*df.Height.std()
upper_limit
```

```python
lower_limit = df.Height.mean() - 3*df.Height.std()
lower_limit
```

`Removing Outliers`

```python
df[(df.Height < lower_limit) | (df.Height > upper_limit)]
```

```python
df_no_outlier_std_dev = df[(df.Height > lower_limit) & (df.Height < upper_limit)]
df_no_outlier_std_dev.head()
```

```python
df_no_outlier_std_dev.shape
```

```python
df.shape
```

#### OUTLIER DETECTION AND REMOVAL USING Z-SCORE

```python
!pip install scikit-learn
```

```python
df.head()
```

```python
df['zscore'] = (df.Height - df.Height.mean()) / df.Height.std()
df.head(5)
```

##### Get data points higher than 3 and lesser than -3

```python
df[df['zscore'] > 3]
```

```python
df[df['zscore'] < -3]
```

```python
df[(df.zscore > 3) | (df.zscore < -3)]
```

#### Removing Outliers

```python
df_no_outliers = df[(df.zscore < 3) & (df.zscore > -3)]
df_no_outliers.head()
```

```python
df_no_outliers.shape
```

```python
df.shape
```

### ENCODING CATEGORICAL VARIABLES

```python
import pandas as pd
df = pd.read_excel ('home prices.xlsx')
df
```

```python
dummies = pd.get_dummies(df.town)
dummies = dummies.astype(int)
print(dummies)
```

```python
merged = pd.concat([df,dummies], axis = 'columns')
merged
```

```python
final_frame = merged.drop(['town', 'Wuse'], axis = 'columns')
final_frame
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
X = final_frame.drop(['price'], axis = 'columns')
X
```

```python
y = final_frame.price
y
```

```python
model.fit(X,y)
```

```python
model.predict([[2800,0,1]])
```

```python
model.predict([[3400,0,0]])
```

```python
model.score(X,y)
```
